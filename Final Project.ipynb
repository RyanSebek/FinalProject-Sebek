```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("spotify.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['emotion'] = le.fit_transform(df['emotion'])
df['Genre'] = le.fit_transform(df['Genre'])
df['Key'] = le.fit_transform(df['Key'])
df['Explicit'] = le.fit_transform(df['Explicit'])
```

```python
df.head(5)
```

```python
numeric_df = df.select_dtypes(include='number',)
rs = np.random.RandomState(42)
corr = numeric_df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
X = df.drop(columns=['artist','song','Genre'])
y = df['Genre']
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
kmeans_orig = KMeans(n_clusters=316, random_state=42)
labels_orig = kmeans_orig.fit_predict(scaled_data)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
kmeans_orig = KMeans(n_clusters=316, random_state=42)
labels_orig = kmeans_orig.fit_predict(scaled_data)
```

```python
from sklearn.metrics import silhouette_score
print("Original K Means Silhouette Score:", silhouette_score(scaled_data, labels_orig))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
```

```python
kmeans = KMeans(n_clusters = 316, random_state=42)
X_pca = pca.fit_transform(scaled_data)
x_clusters = kmeans.fit_predict(X_pca)
```

```python
PCA_Silhouette = silhouette_score(X_pca, x_clusters)
print(f'Silhouette Score: {PCA_Silhouette:.3f}')
```
