```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['emotion'] = le.fit_transform(df['emotion'])
df['Genre'] = le.fit_transform(df['Genre'])
df['Key'] = le.fit_transform(df['Key'])
df['Explicit'] = le.fit_transform(df['Explicit'])
```

```python
df.head(5)
```

```python
numeric_df = df.select_dtypes(include='number',)
rs = np.random.RandomState(42)
corr = numeric_df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
X = df.drop(columns=['artist','song','Genre'])
y = df['Popularity']
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
kmeans_orig = KMeans(n_clusters=316, random_state=42)
labels_orig = kmeans_orig.fit_predict(scaled_data)
```

```python
from sklearn.metrics import silhouette_score
print("Original K Means Silhouette Score:", silhouette_score(scaled_data, labels_orig))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_PCA = pca.fit_transform(scaled_data)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, 'o-', label='Individual Variance')
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 's-', label='Cumulative Variance')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True, linestyle='--', alpha=0.95)
plt.legend()
plt.show()
```
